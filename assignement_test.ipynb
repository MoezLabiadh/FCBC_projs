```python
import pandas as pd
import geopandas as gpd
import folium
```

```python
f = r'\\spatialfiles.bcgov\Work\lwbc\visr\Workarea\moez_labiadh\assignement\WITU_Records_01_12_2022.xlsx'
```

```python
df = pd.read_excel(f, 'WITU Reports')
```

```python
df.head(5)
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Long, df.Lat))
gdf = gdf.set_crs(4326, allow_override=True)
gdf['Date'] = gdf['Date'].astype(str)
```

```python
m = gdf.explore(
         column= 'Number of Birds', # make choropleth based on rhis column
         tooltip= 'Source', # show column value in tooltip (on hover)
         popup=True, # show all values in popup (on click)
         cmap="Dark2",  
         style_kwds=dict(color="gray"), # use black outline
         name='test')

folium.TileLayer('stamenterrain', control=True).add_to(m)
folium.LayerControl().add_to(m)
```
